```python
import matplotlib.pyplot as plt
def f(x: torch.Tensor) -> torch.Tensor:
    return 4 * x ** 2 - torch.sin(5 * x)

def grad_f(x: torch.Tensor) -> torch.Tensor:
    return 12 * x - 5 * torch.cos(5 * x)

x = torch.arange(-1, 1, 0.01)

plt.plot(x, f(x))

x = torch.randn(1)

for _ in range(100):
    x = x - 0.1 * grad_f(x)
    plt.plot(x, f(x), 'ro')

```
